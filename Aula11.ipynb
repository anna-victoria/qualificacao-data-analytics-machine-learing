#Dados de treino e teste

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/victor-ferreira/dataset/main/precificacao_housing_plus.csv")
print(df.shape)
df.head()
```

```python
#criar conjunto de teste
test = df.sample(frac=0.15, random_state=0)

print(test.shape)
test.head()
```

```python
#remover duplicados

df = df.drop(test.index)
df.shape
```

```python
#resetar o index
```

```python
df.reset_index()
test.reset_index()
```
