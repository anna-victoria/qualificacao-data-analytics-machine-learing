#Visualização de dados

##Gráfico de Linhas

```python
#importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importar dados para o df
dados = "https://raw.githubusercontent.com/victor-ferreira/dataset/main/dengue-dataset.csv"
df = pd.read_csv(dados, index_col = "data", parse_dates = True)

#ver as 5 primeiras entradas
df.head()
```

```python
#plotar o num de casos confirmados
fig, ax = plt.subplots()

df['casos-confirmados'].plot(kind="line", ax = ax)
ax.set_title("Dengue em Campinas")
ax.set_ylabel("Casos Confirmados")
ax.set_xlabel("Data")

plt.tight_layout()
##Pesquisar depois esse tight layout
```

```python
#plotar selecionando intervalo
fig, ax = plt.subplots()

df.loc['2002-01-01':'2013-12-31', "temperatura-maxima"].plot(kind="line", ax=ax)
ax.set_title("Temperaturas Máximas em Campinas")
ax.set_ylabel("Temperatura")
ax.set_xlabel("Data")

plt.tight_layout()
```

```python
#plotar selecionando intervalo
fig, ax = plt.subplots()

df.loc['2002-01-01':'2013-12-31', "temperatura-mininima"].plot(kind="line", ax=ax)
ax.set_title("Temperaturas Máximas em Campinas")
ax.set_ylabel("Temperatura")
ax.set_xlabel("Data")

plt.tight_layout()
```

```python
fig, ax = plt.subplots()

df.loc['2002-01-01':'2013-12-31', "chuva"].plot(kind="line", ax=ax)
ax.set_title("Chuva em Campinas")
ax.set_ylabel("Chuva")
ax.set_xlabel("Data")

plt.tight_layout()
```

```python
#plotar vários gráficos
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

#temp média
df.loc['2003-01-01':'2013-12-31', "temperatura-media"].plot(kind="line", ax=ax1)
ax1.set_title("Temperaturas Média em Campinas")
ax1.set_ylabel("Temperatura")
ax1.set_xlabel("Data")

#casos confirmados
df.loc['2003-01-01':'2013-12-31', "casos-confirmados"].plot(kind="line", ax=ax2)
ax2.set_title("Casos de Dengue em Campinas")
ax2.set_ylabel("Casos")
ax2.set_xlabel("Data")

plt.tight_layout()
```

```python
#grafico sobrepostos

#plotar os gráficos normalizados (é meio que deixar um padrão entre os gráficos)
df_normal = (df - df.mean())/df.std()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,10))

#plot temp média
df_normal.loc['2003-01-01':'2013-12-31', "temperatura-media"].plot(kind = "line", ax=ax)
#plot casos confirmados
df_normal.loc['2003-01-01':'2013-12-31', "casos-confirmados"].plot(kind = "line", ax=ax)

#manipular tipos
ax.set_xlabel("Data")
ax.set_title("Temperatura Média e Casos Confirmados")

plt.tight_layout()
```
