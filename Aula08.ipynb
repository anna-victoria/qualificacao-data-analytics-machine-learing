#Visualização de Dados
##Histograma

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = "https://raw.githubusercontent.com/victor-ferreira/dataset/main/dengue-dataset.csv"
df = pd.read_csv(dados)
df.data = pd.to_datetime(df.data)
df.set_index('data', inplace=True)
df.head()
```

```python
df.columns
```

```python
#Histograma
fig,ax = plt.subplots()
ax.hist('temperatura-mininima', data=df)
ax.set_title("Temperatura Mínima")
ax.set_ylabel("Ocorrências")
ax.set_xlabel("Temperatura")

plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.hist('temperatura-mininima', data=df, density=True, bins=30)
ax.set_title("Temperatura Mínima")
ax.set_ylabel("Ocorrências")
ax.set_xlabel("Temperatura")

plt.tight_layout()
```

##Gráfico de Barras

```python
dados1 = "https://raw.githubusercontent.com/victor-ferreira/dataset/main/entrada_passageiro_linha1.csv"
df1 = pd.read_csv(dados1)

df1.head()
```

```python
df1.tail()
```

```python
#apagar linha que há o total
df1.drop(index=23, axis=0, inplace=True)
```

```python
df1.tail()
```

```python
#grafico de barras
fig, ax = plt.subplots(figsize=(10,10))
df1.plot(x='Estação', y='Entradas', kind='bar', ax=ax)
ax.set_title("Entrada Média de Passageiros")
ax.set_xlabel("Estação")
ax.set_ylabel("Entradas")
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
df1.plot(x='Estação', y='Entradas', kind='barh', ax=ax)
ax.set_title("Entrada Média de Passageiros")
ax.set_xlabel("Estação")
ax.set_ylabel("Entradas")
plt.tight_layout()
```

##Gráfico de pizza e donut

```python
dados2 = "https://raw.githubusercontent.com/victor-ferreira/dataset/main/ocorrencias_aviacao.csv"
df2 = pd.read_csv(dados2, index_col='codigo_ocorrencia')

df2.head()
```

```python
#extrair ocorrencias p dicio
ocorrencia_classificacao = dict(df2.ocorrencia_classificacao.value_counts())

#prep dos dados
labels = ocorrencia_classificacao.keys()
sizes = ocorrencia_classificacao.values()
explode=(0.1, 0.1, 0.1)
```

```python
#grafico de pizza
fig, ax = plt.subplots(figsize=(5,5))
ax.pie(sizes, labels=labels)
ax.set_title("Classificação das Ocorrências")
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.pie(sizes, explode = explode, labels=labels, shadow=True)
ax.set_title("Classificação das Ocorrências")
plt.tight_layout()
```

```python
#Grafico de Donut
donut = plt.Circle((0,0), 0.50, fc='white')
fig, ax = plt.subplots(figsize=(5,5))
ax.pie(sizes, labels=labels)
ax.set_title("Classificação das Ocorrências")
ax.add_artist(donut)

plt.tight_layout()
```
